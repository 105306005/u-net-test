```python
from keras.utils import normalize
import os
import glob
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image as im
```

```python
image = im.open("mymask.png")
image_array = np.array(image)

ori_image = im.open("myimage.jpg")
ori_image_array = np.array(ori_image)
# print(ori_image_array)
```

```python
# seg_map - segmentation mask from network, resized_im - your input image

mymask0 = np.equal(image_array, 0) # to get only objects with label 0
mymask0 = np.expand_dims(mymask0, axis=-1) # (H, W) -> (H, W, 1)
mymask0 = np.concatenate((mymask0, mymask0, mymask0), axis=-1) # (H, W, 1) -> (H, W, 3), (don't like it, so if you know how to do it better, please let me know)
mycrops0 = ori_image_array * mymask0 # apply mask on image

mymask1 = np.equal(image_array[:,:], 1) # to get only objects with label 1
mymask1 = np.expand_dims(mymask1, axis=-1) # (H, W) -> (H, W, 1)
mymask1 = np.concatenate((mymask1, mymask1, mymask1), axis=-1) # (H, W, 1) -> (H, W, 3), (don't like it, so if you know how to do it better, please let me know)
mycrops1 = ori_image_array * mymask1 # apply mask on image

mymask2 = np.equal(image_array, 2) # to get only objects with label 2
mymask2 = np.expand_dims(mymask2, axis=-1) # (H, W) -> (H, W, 1)
mymask2 = np.concatenate((mymask2, mymask2, mymask2), axis=-1) # (H, W, 1) -> (H, W, 3), (don't like it, so if you know how to do it better, please let me know)
mycrops2 = ori_image_array * mymask2 # apply mask on image

mymask3 = np.equal(image_array, 3) # to get only objects with label 3
mymask3 = np.expand_dims(mymask3, axis=-1) # (H, W) -> (H, W, 1)
mymask3 = np.concatenate((mymask3, mymask3, mymask3), axis=-1) # (H, W, 1) -> (H, W, 3), (don't like it, so if you know how to do it better, please let me know)
mycrops3 = ori_image_array * mymask3 # apply mask on image

mymask4 = np.equal(image_array, 4) # to get only objects with label 4
mymask4 = np.expand_dims(mymask4, axis=-1) # (H, W) -> (H, W, 1)
mymask4 = np.concatenate((mymask4, mymask4, mymask4), axis=-1) # (H, W, 1) -> (H, W, 3), (don't like it, so if you know how to do it better, please let me know)
mycrops4 = ori_image_array * mymask4 # apply mask on image

plt.figure(figsize=(12, 8))
plt.subplot(231)
plt.title('Original Image')
plt.imshow(ori_image_array)
plt.subplot(232)
plt.title('Testing Label')
plt.imshow(image_array, cmap='jet')
plt.subplot(233)
plt.title('Croped class 0')
plt.imshow(mycrops0)
plt.subplot(234)
plt.title('Croped class 1')
plt.imshow(mycrops1)
plt.subplot(235)
plt.title('Croped class 2')
plt.imshow(mycrops2)
plt.subplot(236)
plt.title('Croped class 3')
plt.imshow(mycrops3)
# plt.subplot(236)
# plt.title('Croped class 4')
# plt.imshow(mycrops4)
plt.show()
```

```python
import cv2
file_name = "mycrops1.png"

src = cv2.imread(file_name, 1)
tmp = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
_,alpha = cv2.threshold(tmp,0,255,cv2.THRESH_BINARY)
b, g, r = cv2.split(src)
rgba = [b,g,r, alpha]
dst = cv2.merge(rgba,4)
cv2.imwrite("mycrops1_test.png", dst)
```
